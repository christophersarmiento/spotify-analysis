```python
#calling import statements

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the .csv files

top10s = pd.read_csv('top2010-2019.csv')
top2017 = pd.read_csv('top2017.csv')
top2018 = pd.read_csv('top2018.csv')
top2019 = pd.read_csv('top2019.csv')
```

```python
#standardizing the speechiness and danceability of top10s and top2019

T10danceList = []
for item in top10s['dnce']:
    item = item/100
    T10danceList.append(item)
    
T10speechList = []
for item in top10s['spch']:
    item = item/100
    T10speechList.append(item)
    
T19danceList = []
for item in top2019['Danceability']:
    item = item/100
    T19danceList.append(item)
    
T19speechList = []
for item in top2019['Speechiness.']:
    item = item/100
    T19speechList.append(item)
```

```python
#plotting each .csv

fig, ax = plt.subplots(ncols=2, nrows =2, figsize=(12,12))

np.linspace(0.0, 1.0, 2, endpoint=True)
T10speech = T10speechList; T17speech = top2017['speechiness']; T18speech = top2018['speechiness']; T19speech = T19speechList; 
T10dance = T10danceList; T17dance = top2017['danceability']; T18dance = top2018['danceability']; T19dance = T19danceList;

ScatT10 = pd.DataFrame({'Speechiness': T10speech ,'Danceability': T10dance })
ScatT10.plot.scatter(x='Danceability', y='Speechiness', title= "Top 10 Songs from 2010-2019", c = 'red', ax=ax[0][0]);

ScatT17 = pd.DataFrame({'Speechiness': T17speech ,'Danceability': T17dance })
ScatT17.plot.scatter(x='Danceability', y='Speechiness', title= "Top Spotify Tracks from 2017", c = 'green', ax=ax[0][1]);

ScatT18 = pd.DataFrame({'Speechiness': T18speech ,'Danceability': T18dance })
ScatT18.plot.scatter(x='Danceability', y='Speechiness', title= "Top Spotify Tracks from 2018", c = 'blue', ax=ax[1][0]);

ScatT19 = pd.DataFrame({'Speechiness': T19speech ,'Danceability': T19dance })
ScatT19.plot.scatter(x='Danceability', y='Speechiness', title= "Top Spotify Tracks from 2019", c = 'orange', ax=ax[1][1]);

plt.sca(ax[0, 0])
plt.xticks(np.arange(0, 1.1, step=0.2))
plt.yticks(np.arange(0, 0.7, step=0.1))
plt.sca(ax[0, 1])
plt.xticks(np.arange(0, 1.1, step=0.2))
plt.yticks(np.arange(0, 0.7, step=0.1))
plt.sca(ax[1, 0])
plt.xticks(np.arange(0, 1.1, step=0.2))
plt.yticks(np.arange(0, 0.7, step=0.1))
plt.sca(ax[1, 1])
plt.xticks(np.arange(0, 1.1, step=0.2))
plt.yticks(np.arange(0, 0.7, step=0.1))

plt.suptitle('Speechiness vs Danceability', fontsize = 14)
```
